```python
from numpy.linalg import svd

B = np.array([[3,2,2], [2,3,-2]])
C = np.array([[2,4], [1,3], [0,0], [0,0]])
```

```python
U,S,VT = svd(B)

print('\nB'); print(B)
print('\nU'); print(U)
print('\nS'); print(np.diag(S, -1)[1:, :])
print('\nV'); print(VT.T)
print('\nU S VT'); print(np.dot(np.dot(U, np.diag(S, -1)[1:, :]), VT))
```

```python
U,S,VT = svd(B, full_matrices=False)

print('\nB'); print(B)
print('\nU'); print(U)
print('\nS'); print(np.diag(S))
print('\nV'); print(VT.T)
print('\nU S VT'); print(np.dot(np.dot(U, np.diag(S)), VT))
```

```python
U,S,VT = svd(C)

print('\nC'); print(C)
print('\nU'); print(U)
print('\nS'); print(np.diag(S, 2)[:, 2:])
print('\nV'); print(VT.T)
print('\nU S VT'); print(np.dot(np.dot(U, np.diag(S, 2)[:, 2:]), VT))
```

```python
U,S,VT = svd(C, full_matrices=False)

print('\nC'); print(C)
print('\nU'); print(U)
print('\nS'); print(np.diag(S))
print('\nV'); print(VT.T)
print('\nU S VT'); print(np.dot(np.dot(U, np.diag(S)), VT))
```

```python
U,S,VT = svd(B, full_matrices=False)

print('[Bv_1 Bv_2]')
print(np.dot(B, VT.T))

print('[s_1u_1 s2_u2]')
print(np.dot(np.diag(S), U.T).T)
```

```python
U,S,VT = svd(C, full_matrices=False)

print('[Cv_1 Cv_2]')
print(np.dot(C, VT.T))

print('[s_1u_1 s2_u2]')
print(np.dot(np.diag(S), U.T).T)
```
